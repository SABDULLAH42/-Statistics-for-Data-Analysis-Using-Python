```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

## Question: Perfume bottles are filled with the average volume of 150 cc and the standard deviation of 2 cc.What percent of bottles will have a volume greater than 153 cc?

```python
norm.sf(153, 150, 2)
```

## Question: Perfume bottles are filled with the average volume of 150 cc and the standard deviation is 2 cc.What percent of bottles will have a volume between 148 and 152 cc?

```python
norm.sf(148, 150, 2) - norm.sf(152, 150, 2)   # the area to the right of 148 ie 84%
                                              # area to the right of 152 is 16%
                                              # So deducting the area 16% (> 152) from 84% we get 68% ie the area b/w 148-152
```

```python
# norm.cdf (x, mu, sigma) - Cumulative distribution function - for less than
# norm.pdf (x, mu, sigma) - Probability density function (not Probability mass function) - for exact
# norm.sf (x, mu, sigma) - for more than (similar to 1-cdf)
```

```python
x_ax = np.linspace (144, 156, 13)   # I want to have 13 items ie value between 144 & 156
```

```python
x_ax
```

```python
y_ax = norm.pdf(x_ax, 150, 2)
```

```python
sns.lineplot(x=x_ax, y=y_ax)       # 
plt.show()
```

```python
x_ax = np.linspace (144, 156, 1300)   # Now we are taking 1300 items for looking the curve more smooth
y_ax = norm.pdf(x_ax, 150, 2)
sns.lineplot(x=x_ax, y=y_ax)
plt.axvline(153, color='red')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('Piece_Dim.csv')
```

```python
df.head()
```

```python
plt.hist(df['Length'])
```

```python
sns.histplot(data=df, x='Length', bins=10)
plt.title('Histogram')
plt.show()
```

```python
sns.displot(data=df, x='Length', bins=10)
plt.title('Histogram')
plt.show()
```

### Now I want to compare the histogram of these two operators

```python
sns.displot(data=df, x='Length', bins=10, hue='Operator')    # hue means color ie what color you want to assign to each operator
plt.title('Histogram')
plt.show()
```

```python
sns.displot(data=df, x='Length', bins=10, col='Operator')
plt.show()
```

# Kde Plot

```python
sns.displot(data=df, x='Length', kind='kde')
```

```python
sns.displot(data=df, x='Length', kind='kde', hue='Operator')
```

```python
sns.displot(data=df, x='Length', kind='ecdf')    # ecdf= empirical cumulative density function
```

```python
sns.displot(data=df, x='Length', kind='hist', rug=True, kde=True)
```

```python
sns.displot(data=df, x='Length', y='Width')    # bivariate plot using 2 variables
```

```python
sns.displot(data=df, x='Length', y='Width', kind='kde')
```

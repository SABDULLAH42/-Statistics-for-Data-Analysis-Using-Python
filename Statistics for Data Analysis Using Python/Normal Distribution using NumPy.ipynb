```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## We are talking about the rod manufacturing where we had the mean as 100 mm and the Standard deviation is 2

```python
np.random.normal(100, 2)
```

```python
np.random.normal(100, 2, 10)            # here 10 is the size of the sample
```

```python
np.round(np.random.normal(100, 2, 10), 2)        # here 2 digit after the decimal and using the round function
```

```python
a = np.random.normal(100, 2 , 1000)
sns.set_theme(style = 'whitegrid')
sns.histplot(x=a)
plt.show()
```

```python
a = np.random.normal(100, 2 , 1000000)
sns.set_theme(style = 'whitegrid')
sns.histplot(x=a, kde=True)
plt.show()
```

```python
sum(a > 98)    # around 84% area is on the right of -1 sigma
```

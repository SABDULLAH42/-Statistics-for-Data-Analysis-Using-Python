# Scipy >>> Stats

```python
#import scipy as sp
Scipy is a huge package we only importing the stats module ie the specific part or module
#import scipy.stats as stats
We are not importing the whole Stats here just importing the piece of the stats module which is binomial distribution
```

```python
from scipy.stats import binom
import numpy as np
```

```python
binom.cdf(2, 20, 0.12)             ## cdf  = cumulative distribution function

#binom.cdf(k, n, p)   where k = number which we are looking for 2 defectives or less or equal to,
                            # n = how many number of pieces we are selecting
                            # p = probabilty here is 12 %
```

# Here we want the probabilty of exactly 2 defectives

```python
binom.pmf(2, 20, 0.12)     ## pmf = probabilty of mass function
```

```python
binom.mean(20, 0.12)
```

```python
binom.std(20, 0.12)
```

```python
binom.var(20, 0.12)
```

# Probability of getting more than 2 defectives

```python
1 - binom.cdf(2, 20, 0.12)
```

```python
binom.sf(2, 20, 0.12)                   ## using binom survival function
```

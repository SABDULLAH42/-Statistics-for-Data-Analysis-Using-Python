```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Piece_Dim.csv')
df.head()
```

```python
plt.plot(df['Length'])
plt.show()
```

```python
plt.plot(df['Length'])
mean_l = df['Length'].mean()                                # mean of length
sd_l = df['Length'].std()
plt.axhline(y=mean_l, color='red', linestyle='-')
plt.axhline(y=(mean_l + 3*sd_l), color='red', linestyle='--')      # mean +3 sigma
plt.axhline(y=(mean_l - 3*sd_l), color='green', linestyle='--')      # mean -3 sigma
plt.show()
```

# Scatterplot

```python
plt.scatter(x = df['Length'], y= df['Width'])
plt.show()
```

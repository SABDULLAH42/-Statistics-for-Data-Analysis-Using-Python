```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Piece_Dim.csv')
```

```python
df.head()
```

```python
sns.relplot(data=df, x='Length', y='Width')       #kind=scatterplot by default in relplot
```

```python
sns.relplot(data=df, x='Length', y='Width', hue='Operator')   #using hue we get the distinction b/w Op-1 and Op-2
```

```python
sns.relplot(data=df, x='Length', y='Width', col='Operator')
```

```python
sns.relplot(data=df, x='Length', y='Width', row='Operator')
```

```python
sns.relplot(data=df, x='Item_No', y='Length', kind='line')
```

```python
sns.relplot(data=df, x='Item_No', y='Length', kind='line', aspect=4)
```

```python
sns.relplot(data=df, x='Item_No', y='Length', kind='line', aspect=4)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.regplot(data=df, x='Length', y='Width')
```

### my_weather DataFrame provides the temperature for 7 days.

Using the .loc method you are required to read the temperature on Monday and assign it to the variable answer.

```python
import pandas as pd
my_weather = pd.DataFrame(
    data=[20, 21, 20, 19, 16, 22, 20], 
    index=['Sunday',  'Monday',  'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'], 
    columns=['Temperature']
)
```

```python
my_weather
```

```python
answer = my_weather.loc[['Monday']]
answer
```

```python
answer = my_weather.loc['Monday'][0]
answer
```

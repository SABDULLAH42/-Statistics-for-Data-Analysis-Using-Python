```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.info()
```

```python
sns.displot(data=titanic, x='age')
```

```python
sns.displot(data=titanic, x='age', hue='sex')
```

### Want to know whether male is more or female

```python
sns.displot(data=titanic, x='age', col='sex')
```

### In that accident, whether more male died or female

```python
sns.displot(data=titanic, x='age', col='sex', row='survived')
```

```python
sns.displot(data=titanic, x='age', col='sex',kind='kde', row='survived')
```

```python
sns.catplot(data=titanic, x='survived', y='age')
```

```python
sns.catplot(data=titanic, x='survived', y='age', kind='swarm', aspect=1.5)
```

```python
sns.catplot(data=titanic, x='survived', y='age', kind='box', aspect=1.5)
```

```python
sns.catplot(data=titanic, x='sex', kind='count')
```

```python
sns.catplot(data=titanic, x='sex', kind='count', hue='survived')
```

```python
sns.catplot(data=titanic, x='deck', kind='count', hue='survived')
```
